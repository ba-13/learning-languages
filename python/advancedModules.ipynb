```python
text = 'My phone number is 408-555-1234'
```

```python
pattern = r'\d\d\d-\d\d\d-\d\d\d\d'
```

```python
match = re.search(pattern, text)
```

```python
match.span()
```

- \+ Occurs one or more times
- {3} Occurs exactly 3 times
- {2,4} Occurs 2 to 4 times
- {3,} Occurs 3 or more times
- \* Occurs zero or more times
- \? Once or None

```python
pattern = r'\d{3}-\d{3}-\d{4}'
```

```python
phonenumber = re.search(pattern, text)
```

```python
phonenumber
```

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})') # makes it seperable. We are still searching the normal way, but this is more flexible.
```

```python
result = re.search(phone_pattern, text)
```

```python
result.group()
```

```python
print(result.group(1))
print(result.group(2))
print(result.group(3))
```

# Timing Python Code

```python
def func(n):
    return [str(i) for i in range(n)]
```

```python
func(10)
```

```python
def func2(n):
    return list(map(str, range(n)))
```

```python
func2(10)
```

```python
import time
```

```python
startTime = time.time()
result = func(100000)
endTime = time.time()
elapsedTime = endTime - startTime
print(elapsedTime)
```

```python
startTime = time.time()
result = func2(100000)
endTime = time.time()
elapsedTime = endTime - startTime
print(elapsedTime)
```

Time it module

```python
import timeit
```

```python
statement1 = '''
func(100)
'''
```

```python
setup1 = '''
def func(n):
    return [str(i) for i in range(n)]
'''
```

```python
timeit.timeit(statement1, setup1, number=100000)
```

```python
statement2 = '''
func2(100)
'''
```

```python
setup2 = '''
def func2(n):
    return list(map(str, range(n)))
'''
```

```python
timeit.timeit(statement2, setup2, number=100000)
```

Timeit only accepts strings as statement (code to execute) and setup (all definitions for that code) so is a bit cumbersome. Using magic of Jupyter, we can use timeit more efficiently.

```python
%%timeit #needs to be the first line of the cell
func2(100)
```

```python
def func2(n):
    return list(map(str, range(n)))
```

```python
%%timeit #needs to be the first line of the cell
func(100)
```

# Zipping and Unzipping Files

```python
with open('file1.txt', 'w+') as f:
    f.write('This is the first file')
```

```python
with open('file2.txt', 'w+') as f:
    f.write('This is the second file')
```

```python
import zipfile
```

```python
comp_file = zipfile.ZipFile('comp_file.zip', 'w')
```

```python
comp_file.write('file1.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('file2.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip', 'r')
```

```python
zip_obj.extractall('extracted_content') # A folder created containing all extracted files.
```

```julia
using Zygote, Plots
using Random
```

```julia
x = collect(1:0.25:100)
y = map(x->3x^2+5x+4, x)
Plots.plot(x, y, seriestype = :scatter)
```

```julia
f(x, w₂, w₁, b) = w₂*x^2 + w₁*x + b
```

```julia
loss(y, t) = (y - t)^2/2
```

```julia
η = 7e-8 # The steps length.
threshold = 1
Random.seed!(1234)
```

```julia
# w₂ⁱ = 4; w₁ⁱ = 6; b = 4;
w₂ⁱ = rand(); w₁ⁱ = rand(); b = rand();

println("(w₂ⁱ, w₁ⁱ, b) ~ ($w₂ⁱ, $w₁ⁱ, $b)")
println("----------------------------------")
grad_w₂_av = 0
grad_w₁_av = 0
grad_b_av = 0
while true
    net_loss = 0
    grad_w₂ = 0
    grad_w₁ = 0
    grad_b = 0
    for i in 1:1:length(x)
        t = f(x[i], w₂ⁱ, w₁ⁱ, b)
        net_loss += loss(t, y[i]) # NetLoss per epoch
        grad_w₂ += - x[i]^2 * (y[i]-t)
        grad_w₁ += - x[i] * (y[i]-t)
        grad_b += - (y[i]-t)
#         println("$i is the counter")
"""
    Updating params per data-item.
#         w₂ⁱ = w₂ⁱ - η*grad_w₂
#         w₁ⁱ = w₁ⁱ - η*grad_w₁
#         b = b - η*grad_b
"""
    end
    grad_w₂_av = grad_w₂/length(x)
    grad_w₁_av = grad_w₁/length(x)
    grad_b_av = grad_b / length(x)
    net_loss = net_loss/length(x)
    w₂ⁱ = w₂ⁱ - η*grad_w₂_av
    w₁ⁱ = w₁ⁱ - η*grad_w₁_av
    b = b - η*grad_b_av
    println("Loss: ", net_loss, "\n")
    
    if net_loss < threshold # Found the minima cost
        break
    end
    if net_loss > 1e10 # Overshooting
        println("Overshoot. Try again")
        break
    end
end
w₂ⁱ, w₁ⁱ, b
```

```julia
Zygote.gradient((x, w₂, w₁, b)->f(x, w₂, w₁, b), 1, 3, 5, 4)
# Output is a tuple, consisting of gradient wrt each corresponding terms in input, at the point provided.
```

```julia
function forward(x, y, w₂, w₁, b)
    ŷ = f(x, w₂, w₁, b)
    l = loss(y, ŷ)
    return l
end
```

```julia
η = 9.8e-8
threshold = 1e15
```

```julia
w₂ = 10*rand(); w₁ = 10*rand(); b = 10*rand();
println("(w₂, w₁, b) = ($w₂, $w₁, $b)");
println("------------------------------------")
net_loss_r = 0
while true
    net_loss = 0
    grad_w₂ = 0
    grad_w₁ = 0
    grad_b = 0
    for i = 1:length(x)
        net_loss += forward(x[i], y[i], w₂, w₁, b)
        _,_,g_w₂,g_w₁,g_b = Zygote.gradient((a,b,c,d,e)->forward(a,b,c,d,e),x[i],y[i],w₂,w₁,b)
        grad_w₂ += g_w₂
        grad_w₁ += g_w₁
        grad_b += g_b
    end
    grad_w₂_av = grad_w₂/length(x)
    grad_w₁_av = grad_w₁/length(x)
    grad_b_av = grad_b/length(x)
    w₂ = w₂ - η*grad_w₂_av
    w₁ = w₁ - η*grad_w₁_av
    b = b - η*grad_b_av

    println("Loss: ", net_loss)
    if net_loss < 0.5
        break
    end
    if net_loss > threshold
        break
    end
    net_loss_r = net_loss
end
println("(w₂, w₁, b) = ($w₂, $w₁, $b)");
```
